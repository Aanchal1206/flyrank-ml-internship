## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*



**Task Type:** Classification

The Content Refreshing lane is best framed as a **classification** problem. The goal is to predict whether a content page should be refreshed or not based on historical performance signals. The model outputs one of two classes:

- **1:** Needs Refresh
- **0:** Does Not Need Refresh

This classification helps content teams identify which pages require updates without manually reviewing every page.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*



**Target:** `needs_refresh`

The target is a binary label indicating whether a content page requires refreshing.

If historical labels are available, they can come from observed editorial decisions where pages were refreshed successfully. If such labels are unavailable, a proxy label can be created using business rules, such as low traffic, declining engagement, or outdated content.

Example:

- 1 = Needs Refresh
- 0 = Does Not Need Refresh

## 3. Success metric

*One metric you can defend. What number means 'good'?*



**Metric:** F1-Score

F1-Score is appropriate because both false positives and false negatives are important.

- A false positive wastes editorial effort by refreshing content unnecessarily.
- A false negative misses content that should have been updated.

A good model should achieve a high F1-score while balancing precision and recall.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*



A fixed rule such as "refresh every page older than one year" would ignore many important signals.

Some older pages continue to perform well, while newer pages may lose visibility because of changing search trends or user behavior.

Machine learning can analyze multiple factors together, including historical search performance, trend direction, engagement-related signals, and other page characteristics. By learning patterns from historical data, the model can better prioritize which pages should be reviewed than a single fixed rule.

The model supports editorial decision-making rather than replacing human judgment.
